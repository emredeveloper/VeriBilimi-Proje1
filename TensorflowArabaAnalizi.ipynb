```python
dataFrame = dataFrame.drop("transmission", axis = 1)
```

```python
y = dataFrame["price"].values
x = dataFrame.drop("price" ,axis = 1).values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(dataFrame)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #modelimizi oluşturoyuruz
from tensorflow.keras.layers import Dense # katmanlarımızı oluşturuyoruz
```

```python
x_train.shape
```

```python
model = Sequential() # modelimizi oluşturduk
```

```python
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1)) # çıkış katmanı chatgpt = Bu, modelin son katmanı olacaktır ve modelin tahminlerini vermesi için kullanılacaktır.
#Bu katmanın aktivasyon fonksiyonu belirtilmemiştir, bu nedenle varsayılan olarak lineer aktivasyon fonksiyonu kullanılır.

model.compile(optimizer="adam",loss="mse")
# birçok durumda adam daha iyi perfonmans gösteriyor
```

```python
model.fit(x=x_train, y = y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi
```

```python
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
